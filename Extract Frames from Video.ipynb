# Python code to get frames from video

```python
import cv2
import math

cap = cv2.VideoCapture("DATA/testvid.mp4")
frameRate = cap.get(5) #frame rate
while(cap.isOpened()):
    frameId = cap.get(1) #current frame number
    ret, frame = cap.read()
    if (ret != True):
        break
    if (frameId % math.floor(frameRate) == 0):
        filename = "OUTPUT/image_" +  str(int(frameId)) + ".jpg"
        cv2.imwrite(filename, frame)
cap.release()


```

```python
import cv2

def one_image_per_second(video_file, output_dir, fps):
   cap = cv2.VideoCapture(video_file)
   frame_num = 0
   image_num = 0
   while cap.isOpened():
       ret, frame = cap.read()
       if ret:
           if frame_num % fps == 0:
               cv2.imwrite(output_dir+"/image{}.jpg".format(image_num),frame)
               image_num += 1
       else:
           break
       frame_num += 1

   cap.release()
   cv2.destroyAllWindows()
   return

one_image_per_second("DATA/testvid.mp4", "OUTPUT",30)
```
